# מחברת לחקר ועיבוד נתוני תאונות דרכים
מחברת זו משמשת לטעינה, הצגה, וניקוי ראשוני של המערך במטרה להכינו לאימון מודלים (Machine Learning).

### משימה 1: טעינת נתונים והצגת דוגמאות

```python
import pandas as pd

# טעינת הקובץ לתוך משתנה (Dataframe)
file_path = 'data/israel_road_accidents_simulated.csv'
df = pd.read_csv(file_path)

# הצגת 5 השורות הראשונות
display(df.head())
```

### משימה 2: מספר שורות, עמודות וסוג כל עמודה

```python
print(f"מספר שורות: {df.shape[0]}")
print(f"מספר עמודות: {df.shape[1]}\n")

print("=== סוגי העמודות ===")
print(df.dtypes)
```

### משימה 3: ניקוי ערכים חסרים מותנה
1. בדיקת חסרים לפני ניקוי
2. מחיקת עמודות עם מעל 50% חסרים
3. מילוי ערכים (ממוצע בקטגוריה מספרית, ערך שכיח בקטגוריה טקסטואלית)

```python
# --- תצוגת סיכום *לפני* ---
print("=== השורות/העמודות החסרות בלבד (תוצאה ריקה משמעותה 0 חסרים) ===")
missing_before = df.isnull().sum()
print(missing_before[missing_before > 0]) 
print(f"\nסה'כ שורות ועמודות לפני הניקוי: {df.shape}\n")

# 2. חישוב אחוז החסרים לכל עמודה
missing_percentage = df.isnull().mean() * 100

# 3. מחיקת עמודות עם מעל 50% ערכים חסרים
cols_to_drop = missing_percentage[missing_percentage > 50].index
df = df.drop(columns=cols_to_drop)
if len(cols_to_drop) > 0:
    print(f"העמודות שנמחקו כי הכילו מעל 50% חסרים: {list(cols_to_drop)}\n")

# 4. מילוי ערכים חסרים בעמודות שנותרו
for col in df.columns:
    # בדיקה האם אכן חסרים ערכים (אם יש 0 זה אפילו לא נכנס לפעולות המילוי)
    if df[col].isnull().sum() > 0:
        # בדיקה אם העמודה היא נומרית (מספרים)
        if pd.api.types.is_numeric_dtype(df[col]):
            mean_val = df[col].mean()
            df[col] = df[col].fillna(mean_val)
        # עמודה קטגוריאלית (טקסט/אובייקט) - מילוי עפ"י הערך השכיח
        else:
            mode_val = df[col].mode()[0]
            df[col] = df[col].fillna(mode_val)

# --- תצוגת סיכום *אחרי* ---
print("=== סיכום ערכים חסרים אחרי הניקוי ===")
missing_after = df.isnull().sum()
print(missing_after) 
print(f"\nסה'כ שורות ועמודות אחרי הניקוי: {df.shape}")
```
